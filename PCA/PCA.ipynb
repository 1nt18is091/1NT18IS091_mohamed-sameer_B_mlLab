```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy.special import expit
from numpy.linalg import eig
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,1]
y=df.iloc[:,2]
```

```python
x_mean=0;
y_mean=0;
n=len(x)

for i in range(n):
    x_mean+=x[i]
    y_mean+=y[i]
    
x_mean=x_mean/n
y_mean=y_mean/n
```

```python
covar_xy=0;
covar_xx=0;
covar_yy=0
for i in range(n):
    covar_xy+=((x[i]-x_mean)*(y[i]-y_mean))
    covar_xx+=((x[i]-x_mean)*(x[i]-x_mean))
    covar_yy+=((y[i]-y_mean)*(y[i]-y_mean))

covar_xy=covar_xy/(n-1)
covar_xx=covar_xx/(n-1)
covar_yy=covar_yy/(n-1)
```

```python
C=[[covar_xx,covar_xy],[covar_xy,covar_yy]]
C
```

```python
values,vectors=eig(C)
```

```python
max_values=values[0];
max_i=0;
for i in range(len(values)):
    if(max_values<values[i]):
        max_i=i
        max_values=values[i]
eig_value=values[max_i]
eig_vector=vectors[:,max_i]
V=eig_vector
```

```python
Y=[]

for i in range(n):
    Y.append((x[i]*V[0])+(y[i]*V[1]))
red=pd.DataFrame(Y)
red
```
